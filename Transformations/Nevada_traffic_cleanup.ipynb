```python
# Drop Global UTC-Time column
clark_co_df.drop(columns = ['CrashDateTimeUTC','Distance','GlobalID'], inplace=True)
clark_co_df.columns
```

```python
unk_cols = ['V1_Driver_Factors','V1_Driver_Distracted','V1_Vehicle_Factors','V1_Most_Harmful_Event',
            'V1_All_Events','V2_Type','V2_Action','V2_Driver_Factors',
            'V2_Driver_Distracted','V2_Vehicle_Factors','V2_Most_Harmful_Event','V2_All_Events',
            'First_Harmful_Event','Nonmotorist_Factors','Factors_Roadway','Lighting',
            'HWY_Factors']

for x in unk_cols:
    clark_co_df[x].fillna('UNKNOWN', inplace=True)

clark_co_df
```

```python
zerolist = ['Fatalities','Injured','V1_Lane_Num','V2_Lane_Num']
for zeros in zerolist:
    clark_co_df[zeros].fillna(0, inplace=True)

clark_co_df
```

```python
nolist = ['Property_Damage_Only','Pedalcyclist','Pedestrian','Motorcyclist','AnimalType']

for nos in nolist:
    clark_co_df[nos].fillna('NO', inplace=True)

clark_co_df
```

```python
ulist = ['V1_Dir','V2_Dir','Injury_Type']

for us in ulist:
    clark_co_df[us].fillna('U', inplace=True)

clark_co_df
```

```python
# Fill in empty age columns with median 
clark_co_df['V1_Driver_Age'] = clark_co_df['V1_Driver_Age'].fillna(clark_co_df['V1_Driver_Age'].mean())
clark_co_df['V2_Driver_Age'] = clark_co_df['V2_Driver_Age'].fillna(clark_co_df['V2_Driver_Age'].mean())
```

```python
clark_co_df['Fatalities'] = clark_co_df['Fatalities'].astype(int)
clark_co_df['Injured'] = clark_co_df['Injured'].astype(int)

clark_co_df['Accident_Rec_Num'] =  clark_co_df['Accident_Rec_Num'].astype(int)
```

```python
# Fill in blank age columns with median; recast all ages as integers
clark_co_df['V1_Driver_Age'] = clark_co_df['V1_Driver_Age'].astype(int)
clark_co_df['V2_Driver_Age'] = clark_co_df['V2_Driver_Age'].astype(int)
```

```python
clark_co_df['Crash_Type'].replace(to_replace="SIDESWIPE - MEETING", value = "SIDESWIPE, MEETING", inplace=True)
clark_co_df
```

```python
clark_co_df['Factors_Roadway'].replace(to_replace=["ICE: SNOW","ICE: WET", "OTHER: ICE: SNOW"], value = "ICE", inplace=True)
clark_co_df
```

```python
clark_co_df['Injury_Type'].replace('U', 'UNKNOWN', inplace=True)
clark_co_df['Injury_Type'].replace('K', 'FATAL', inplace=True)
clark_co_df['Injury_Type'].replace('A', 'SUSPECTED SERIOUS INJURY', inplace=True)
clark_co_df['Injury_Type'].replace('B', 'SUSPECTED MINOR INJURY', inplace=True)
clark_co_df['Injury_Type'].replace('C', 'POSSIBLE INJURY', inplace=True)
```

```python
clark_co_df['Lighting'].replace(to_replace='DARK- SPOT LIGHTING',value = 'DARK - SPOT LIGHTING', inplace=True)
```
