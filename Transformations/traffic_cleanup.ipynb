```python
#import dependacies
import pandas as pd 
```

```python
# Read in data
raw_traffic_df = pd.read_csv('../Resources/Traffic_Crashes.csv')
raw_traffic_df
```

```python
# Columns to drop based off group notes
drop_cols = ['ACRES','Agency','AREA',
            'CNTY_IS','County','Crash_Time',
            'CrashDateT','PERIMETER','SHADE','SQ_MILES','WARD']

raw_traffic_df.drop(columns = drop_cols, inplace=True)
raw_traffic_df
```

```python
#Check the null values
print(raw_traffic_df.isnull().sum())
```

```python
# View columns
raw_traffic_df.columns
```

```python
# # Drop columns that have more than half the data missing
# dumb_columns = []
# for col in raw_traffic_df.columns:
#     if raw_traffic_df[col].isnull().sum() > 30000:
#         dumb_columns.append(col)

# raw_traffic_df.drop(columns = dumb_columns, inplace=True)
new_traffic_df = raw_traffic_df.copy()
```

```python
# Replace null values with 0 in the Injured and Fatalities column

new_traffic_df["Fatalities"].fillna(0, inplace=True)
new_traffic_df["Injured"].fillna(0, inplace=True)
```

```python
# Check to see if solution worked
new_traffic_df.isnull().sum()
```

```python
# Check the last few columns
new_traffic_df.iloc[:,18:].isnull().sum()
```

```python
#Look at columns to work with
new_traffic_df.columns
```

```python
#Replace null values with 'Unknown
new_traffic_df['V1_Vehicle'].fillna('Unknown', inplace=True)
new_traffic_df['V1_Type'].fillna('Unknown', inplace=True)
new_traffic_df['V1_Driver1'].fillna('Unknown', inplace=True)
new_traffic_df['V2_Type'].fillna('Unknown', inplace=True)
new_traffic_df['V2_Action'].fillna('Unknown', inplace=True)
new_traffic_df['V2_Driver1'].fillna('Unknown', inplace=True)
new_traffic_df['Factors_Ro'].fillna('Unknown', inplace=True)
new_traffic_df['Lighting'].fillna('Unknown', inplace=True)
new_traffic_df['HWY_Factor'].fillna('Unknown', inplace=True)

```

```python
# Check to verify my solution worked
new_traffic_df.isnull().sum()
```

```python
# Replace null values with 0
new_traffic_df['V1_Lane_Nu'].fillna(0, inplace=True)
new_traffic_df['V1_Driver_'].fillna(0, inplace=True)
new_traffic_df['V2_Lane_Nu'].fillna(0, inplace=True)
new_traffic_df['V2_Driver_'].fillna(0, inplace=True)
```

```python
# Check to see if solution worked
new_traffic_df.isnull().sum()
```

```python
# Replace null values with 'U' for Unknown
new_traffic_df['V1_Dir'].fillna('U', inplace=True)
new_traffic_df['V2_Dir'].fillna('U', inplace=True)
new_traffic_df['Dir'].fillna('U', inplace=True)
```

```python
new_traffic_df.isnull().sum()
```

```python
# Drop Crash time since this is not correct data
new_traffic_df = new_traffic_df.drop(columns = ['Crash_Time'])
new_traffic_df.head()
```

```python
# Change the Crash date to Date time
new_traffic_df['Crash_Date'] = pd.to_datetime(new_traffic_df['Crash_Date'], format = '%Y-%m-%d %H:%M')
new_traffic_df.dtypes
```
